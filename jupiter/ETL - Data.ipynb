```python
import pandas as pd
%load_ext autoreload
%autoreload 2
import os
import utils
import warnings
warnings.filterwarnings("ignore")
```

```python
# Ruta de la carpeta que contiene los archivos
ruta_carpeta = '../data'

# Lista para almacenar DataFrames de cada archivo
dataframes = []

# Iterar a través de cada archivo en la carpeta
for archivo in os.listdir(ruta_carpeta):
    if archivo.endswith('.csv'):
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        # Leer el archivo CSV y añadir el DataFrame a la lista
        dataframes.append(pd.read_csv(ruta_archivo))

# Concatenar todos los DataFrames en un solo DataFrame
df_data = pd.concat(dataframes, ignore_index=True)

```

```python
# Elimino columnas no necesarias o de alto % nulo
columnas_a_borrar = ['titular_pais_nacimiento_id', 'titular_domicilio_provincia_id', 'titular_pais_nacimiento_indec_id', 'titular_domicilio_provincia_indec_id','fecha_inscripcion_inicial','automotor_tipo_codigo',]

# Eliminar las columnas especificadas
df_data = df_data.drop(columnas_a_borrar, axis=1)
```

```python
utils.types_df(df_data)
```

```python
# Convertir la columna 'tramite_fecha' a datetime
df_data['tramite_fecha'] = pd.to_datetime(df_data['tramite_fecha'], errors='coerce')
```

```python
df_data.to_csv('../data/data.csv', index=False)
```
